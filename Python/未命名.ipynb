# 中文变阿拉伯数字

```python
'''
中文计数：数字 + 单位
万以及万以下： 直接单位乘以前面的那个数字就好

亿到万或者超过亿：不能只是单纯的单位乘以前面的数字，还需要再乘上万或者亿

'''



# 数字映射
number_map = {
    "零": 0,
    "一": 1,
    "二": 2,
    "三": 3,
    "四": 4,
    "五": 5,
    "六": 6,
    "七": 7,
    "八": 8,
    "九": 9
}

# 单位映射
unit_map = {
    "十": 10,
    "百": 100,
    "千": 1000,
    "万": 10000,
    "亿": 100000000
}

def backward_cn2an_three(inputs):
    output = 0
    unit = 1
    # 万、亿的单位
    ten_thousand_unit = 1
    num = 0
    for index, cn_num in enumerate(reversed(inputs)):
        if cn_num in number_map:
            # 数字
            num = number_map[cn_num]
            # 累加
            output = output + num * unit
        elif cn_num in unit_map:
            # 单位
            unit = unit_map[cn_num]
            # 判断出万、亿
            if unit % 10000 == 0:
                # 万、亿
                if unit > ten_thousand_unit:
                    ten_thousand_unit = unit
                # 万亿
                else:
                    ten_thousand_unit = unit * ten_thousand_unit
                    unit = ten_thousand_unit

            if unit < ten_thousand_unit:
                unit = ten_thousand_unit * unit
        else:
            raise ValueError(f"{cn_num} 不在转化范围内")

    return output



res = backward_cn2an_three("一百二十三")
print(res)
res = backward_cn2an_three("一千二百三十四万五千六百七十八")
print(res)
res = backward_cn2an_three("一亿二千三百四十五万六千七百八十一")
print(res)
res = backward_cn2an_three("一千二百三十四万五千六百七十八亿一千二百三十四万五千六百七十八")
print(res)

'''
123
12345678
123456781
1234567812345678
'''
```

```python
def eraseOverlapIntervals(intervals):
    if intervals == [[]] or intervals == []:
        return 0
    
    intervals.sort(key = lambda x:x[1])
    res = [intervals[0]]
    #count = 0
    overlap = 0
    
    for start, end in intervals[1:]:
        pre_end = res[-1][1]
        if start <= pre_end:
            if start <= overlap:
                pass
            else:
                overlap = pre_end
                res.append([start, end])
        else:
            res.append([start, end])
    return len(intervals) - len(res)


res = eraseOverlapIntervals([[0, 3], [1, 3], [2, 3], [4, 5]])
print(res)
res = eraseOverlapIntervals([[1,2], [1,2], [1,2]])
print(res)
res = eraseOverlapIntervals([[1,2], [2,3]])
print(res)
res = eraseOverlapIntervals([[1,2], [2,3], [3,4], [1,3]])
print(res)
res = eraseOverlapIntervals([[0,3], [2,6], [5,8], [7,10]])
print(res)
res = eraseOverlapIntervals([[0,3], [2,6], [5,8], [0,10]])
print(res)
```

```python
def isOverlap(interval1, interval2):
    if interval1[1] >= interval2[0]:
        return [interval2[0], interval1[1]]
    else:
        return None

def removeInterval(intervals):
    # 按照区间末端排序
    intervals = sorted(intervals, key=lambda x: (x[1]))
    # 记录前一个区间的重叠状况
    prev = (intervals[0], None)
    # 待删除的区间计数
    count = 0
    # 遍历每一个区间
    for i in range(1, len(intervals)):
        # 计算当前区间和上一个区间是否有重叠
        overlap = isOverlap(prev[0], intervals[i])
        # 如果没有重叠，用prev记录状况，去看下一个区间
        if not overlap:
            prev = (intervals[i], None)
        # 如果和上一个区间有重叠
        else:
            # 如果上一个区间和前面的区间无重叠
            if not prev[1]:
                # 用prev记录状况，去看下一个区间
                prev = (intervals[i], overlap)
                # 如果上一个区间和前面的区间有重叠
            else:
            # 需要看两个重叠区间是否有重叠，如果有，那么当前区间就要被删除
                if isOverlap(prev[1], overlap):
                    count += 1
                # 如果没有，那么无所谓，用prev记录状况，去看下一个区间
                else:
                    prev = (intervals[i], overlap)
    # 返回计数结果
    return count


res = removeInterval([[0, 3], [1, 3], [2, 3], [4, 5]])
print(res)
res = removeInterval([[1,2], [1,2], [1,2]])
print(res)
res = removeInterval([[1,2], [2,3]])
print(res)
res = removeInterval([[1,2], [2,3], [3,4], [1,3]])
print(res)
res = removeInterval([[0,3], [2,6], [5,8], [7,10]])
print(res)
res = removeInterval([[0,3], [2,6], [5,8], [0,10]])
print(res)
```
