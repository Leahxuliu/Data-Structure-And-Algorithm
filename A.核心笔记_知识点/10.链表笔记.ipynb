# 知识点

## 定义
* Linklist ListNode
* 链表由一系列结点（链表中每一个元素称为结点）组成，结点可以在运行时动态生成。 每个结点包括两个部分：一个是存储数据元素的数据域，另一个是存储下一个结点地址的指针域。
* 类似单边树
* 写成1->2->3->4->5

```python
class ListNode:
    def __init__(self, value):
        self.val = value
        self.next = None
```

    * 前后段对比

* 转置法（难点）
    * 加和
        * 1->2->3 代表123
        * 4->5->6 代表456
        * 两者相加得到5->7->9
            * 方法一
                * 先把两个链表转置
                * 用分离指针两两相加，再转置
            * 方法二
                * 把链表变成数字，再把数字变成链表
    * 回文判断
        * 把原有的链表转置，形成一个新的链表
        * 用分离指针进行判断两者是否相同
    * 交叉合并
        * 例如 1->2->3->4->5 中点前的链表转置之后插入中点后的链表中
        * 得到 4->3->5->2->1
            * s,f找中点
            * 把前半段转置插入后半段

```python
class Linklist:
    def reverse(self, head):  # reverse the linkedlist,and return new head
        if not head:
            return head
        
        curr = None
        while head:
            nextHead = head.next
            head.next = curr
            curr = head
            head = nextHead
        return curr
```

**转list/set/dic/数字法**   
* 转成list，利用list的性质，对元素进行操作
    * 优点：方便，简单
    * 缺点：
        * 占据空间，空间复杂度为O(N)
        * a in list，时间复杂度为O(N)
* 转成set
    * 优点：a in set，时间复杂度为O(1)
    * 缺点：set.add()，之后set里元素顺序并不是添加顺序
* 转成dic
    * 优点：
        * a in dic，时间复杂度为O(1)
        * 记录node出现次数
        * 记录复杂链表的对应关系
    * 缺点
        * 无切片等操作
* 转成数字
    * 优点：算数运算方便
    * 缺点：链表和数字相互转换，时间复杂度为O(N)

```python
class Linklist:
    # 链表转数字，注意链表和数字位数的对应关系，是不是逆序
    def transferInteger(self, head):  # transfer 1->2->3 to 321
        if not head:
            return 0
        
        res = 0
        times = 1
        while head:
            res = res + head.val * times
            times *= 10
        return res
```

```python
class Linklist:
    # 数字转链表，采用str转化数字，注意位数的对应关系；single digit个位，tens digit十位
    def transferLinklist(self, numb):  # transfer numb to linkedlist, 321 -> 1 ->2 ->3
        dummy = curr = ListNode(-1)
        for val in str(numb)[::-1]:
            curr.next = ListNode(int(val))
            curr = curr.next
        return dummy.next
```
