# 100 - 相同的树



```python
#方法一 递归
# (DFS pre-order)
# 用递归的方法对根和左右进行判断
# using recursion to check the root, left-child and right-child

class Solution:
    def isSameTree(self, p, q):
        """
        :type p: TreeNode
        :type q: TreeNode
        :rtype: bool
        """    
        # p and q are both None
        if not p and not q:  # p 和 q都已经结束了，没有节点了；或者，这两个树，原本就是空的
            return True
        # one of p and q is None
        if not q or not p:  # p或q，其中一个已经遍历完了，没有节点了；或者，其中一个树，原本就是空的
            return False
        if p.val != q.val:  # 有节点，但是值不同
            return False
        return self.isSameTree(p.left, q.left) and \
               self.isSameTree(p.right, q.right)

'''
最开始
p: TreeNode{val: 1, left: TreeNode{val: 2, left: None, right: None}, right: TreeNode{val: 3, left: None, right: None}}
q: TreeNode{val: 1, left: TreeNode{val: 2, left: None, right: None}, right: TreeNode{val: 3, left: None, right: None}}

p与p.val的区别
p是TreeNode
p.val是值，比如1，2，3

'''
```

```python
# 迭代（BFS）
# 建立一个flag_all
# 注意None的情况
# 自己写的，写法有点繁琐

from collections import deque
class Solution:
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        if not p and not q:
            return True
        if not p or not q:
            return False
        
        flag_all = True
        
        p_queue = deque()
        q_queue = deque()
        
        p_queue.append(p)
        q_queue.append(q)
        
        while p_queue:
            p_node = p_queue.popleft()
            q_node = q_queue.popleft()
            
            if p_node.val != q_node.val:
                flag_all = False
                break
            else:
                if p_node.left and q_node.left:
                    p_queue.append(p_node.left)
                    q_queue.append(q_node.left)
                    
                if p_node.right and q_node.right:
                    p_queue.append(p_node.right)
                    q_queue.append(q_node.right)
                    
                if p_node.left and not q_node.left:
                    flag_all = False
                    break
                elif not p_node.left and q_node.left:
                    flag_all = False
                    break
                elif p_node.right and not q_node.right:
                    flag_all = False
                    break
                elif not p_node.right and q_node.right:
                    flag_all = False
                    break
                    
        if flag_all == True:
            return True
        else:
            return False
```

```python
# BFS的优化写法
# q和p 写在一个duque里面

from collections import deque
class Solution:
    def isSameTree(p: TreeNode, q: TreeNode) -> bool:
    
        queue = deque([(p, q)])
        
        while queue:
            p_node, q_node = queue.popleft()

            if not p_node and not q_node:
                return True
            
            elif not p_node or not q_node:
                return False
            
            elif p_node.val != q_node.val:
                return False

            else:
                queue.append((p_node.left, q_node.left))
                queue.append((p_node.right, q_node.right))
        return True
```

```python
'''
    1
   / \
  2   3
 / \ / \
4  5 6  7

[2,3] [4,7] [5,6] 两两一对
4：2.left
7: 3.right


思路：recursion based on pre-order dfs
time complex:  O(N)
space complex: worst O(N), average(logN)
'''

from collections import deque

class Solution(object):
    def isSymmetric(self, root):
        if root == None:
            return True
        
        queue = deque()
        queue.append([root.left, root.right])
        
        while queue:
            n1, n2  = queue.popleft()
            
            if n1 == None and n2 == None:
                pass
            elif n1 == None or n2 == None:
                return False
            elif n1 != None and n2 != None:
                if n1.val != n2.val:
                    return False
                else:
                    queue.append([n1.right, n2.left])
                    queue.append([n1.left, n2.right])
        return True
                
'''
易错点：
if n1 == None and n2 == None:
    pass
    
如果n1 == None and n2 == None 就return Ture的话
[9,-42,-42,null,76,76,null,null,13,null,13]
就是错的

'''
```

# 104 - 二叉树的最大深度

## 题目理解
1. 给定一个二叉树，找出其最大深度。
2. 二叉树的深度为根节点到最远叶子节点的最长路径上的节点数。
3. 说明：叶子节点是指没有子节点的节点。

## 思路
方法1: 迭代（BFS）   
* 树为空，return 0
* 累计depth, 在queue里面构建[node, depth],这里的depth是指，这个node所对应的depth
* 将depth与max_depth进行比较，更新最大深度，这里的max_depth是指的整体树的最大深度

方法2: 递归（DFS）    
* 

```python
'''
time complex: O(N)
space complex: best(1), average O(N)
'''

from collections import deque
class Solution:
    def maxDepth(self, root: TreeNode) -> int: 
        if root == None:
            return 0
        
        max_depth = 1  # 只有根的时候，depth是1
        queue = deque()
        queue.append([root, max_depth])
        
        while queue:
            node, depth = queue.popleft()
            max_depth = max(max_depth, depth)
            if node.left != None:
                queue.append([node.left, depth+1])
            if node.right != None:
                queue.append([node.right, depth+1])
            
        return max_depth
```

```python
'''
time complex: O(N)
space complex: worst O(N), average O(logN)

'''
class Solution:
    def maxDepth(self, root: TreeNode) -> int: 
        
        if root == None:
            return 0

        # 叶子节点，返回1
        if root.left == None and root.right == None:
            return 1

        # 当左右孩子至少有一个不为空时，返回左右孩子较大深度的节点值
        if root.left != None or root.right != None:
            return 1+max(self.maxDepth(root.left),self.maxDepth(root.right))
```

# 1361. Validate Binary Tree Nodes

题目理解：   
1. 一个数，一个root，一个leftChild，一个rightChild
2. 有一个n个node的树，node numbered from 0 to n - 1
3. 如果第i个node没有left child， leftChild[i] == -1
4. where node i has two children leftChild[i] and rightChild[i], return true


Constraints:   
* 1 <= n <= 10^4
* leftChild.length == rightChild.length == n
* -1 <= leftChild[i], rightChild[i] <= n - 1

```python
'''
思路：
跟DFS，BFS都没有关系！
画图找规律！

n: node的个数

按照题目的要求，构成一个树，len(leftChild) + len(rightChild) == 2n
-1也就是None的个数，一定是 n + 1

'''

class Solution:
    def validateBinaryTreeNodes(self, n: int, leftChild: List[int], rightChild: List[int]) -> bool:
        if len(leftChild) == len(rightChild) == n:
            if leftChild.count(-1) + rightChild.count(-1) == n + 1:
                return True
        else:
            return False
```

# 662.Maximum Width of Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

from collections import deque
class Solution:
    def widthOfBinaryTree(self, root: TreeNode) -> int:
        if root == None:
            return 0
        
        queue = deque()
        queue.append([root, 1])
        
        max_width = 0
        
        while queue:
            queue_len = len(queue)
            min_index = queue[0][1]
            max_index = queue[-1][1]
            
            for i in range(queue_len):
                node, index = queue.popleft()
                
                if node.left != None:
                    queue.append([node.left, 2*index])

                if node.right != None:
                    queue.append([node.right, 2*index + 1])
                
            max_width = max(max_width, (max_index - min_index + 1))
            
        return max_width
        
```

# 144
**题目**   
 preorder traversal 
 
 **思路**  
 用dfs的preorde遍历
 
 

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        nodes = []
        
        if root == None:
            return nodes
        self.dfs(nodes, root)
        return nodes
        
    def dfs(self, nodes, root):
        nodes.append(root.val)
        if root.left != None:
            self.dfs(nodes, root.left)
        if root.right != None:
            self.dfs(nodes, root.right)
```

```python
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        nodes = []
        
        if root == None:
            return nodes
        
        self.dfs(nodes, root)
        return nodes
    
    def dfs(self, nodes, root):
        if root.left != None:
            self.dfs(nodes, root.left)
        
        nodes.append(root.val)
        
        if root.right != None:
            self.dfs(nodes, root.right)
```
